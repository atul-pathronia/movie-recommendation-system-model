```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
movies = pd.read_csv(r"C:\Users\hp\Desktop\Data-Analyst\Movie-Recommendation-System/TMDB_movie_dataset_v11.csv")
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.shape
```

```python
movies.duplicated().sum()
```

```python
movies["keywords"] = movies["keywords"].str.split(",")
```

```python
movies["keywords"] = movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies["genres"] = movies["genres"].str.split(",")
```

```python
movies["genres"] = movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies["overview"] = movies["overview"].str.split(",")
```

```python
movies["tags"] = movies["overview"] + movies["keywords"] +  movies["genres"]
```

```python
movies.head()
```

```python
new_movies = movies[["id","title","tags","backdrop_path"]]
```

```python
new_movies["tags"] = new_movies["tags"].apply(lambda x: " ".join(x))
```

```python
new_movies["tags"] = new_movies["tags"].apply(lambda x: x.replace("\'",""))
```

```python
new_movies["tags"] = new_movies["tags"].apply(lambda x: x.lower())
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_movies["tags"]  = new_movies["tags"].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words="english")
```

```python
vectors = cv.fit_transform(new_movies["tags"]).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_matrix = cosine_similarity(vectors)
```

```python
def recommender_func(movie):
    try:
        movie_index = new_movies[new_movies["title"] == movie].index[0]
        distance = similarity_matrix[movie_index]
        movies_list = sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
        for i in movies_list:
            print(new_movies.iloc[i[0]].title)
    except:
        print("Error")
```

```python
recommender_func("The Avengers")
```
